```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  

from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets.samples_generator import make_circles
import numpy as np
```

```python
# factor adalah digunakan untuk scale jarak yang dipusat dan dipinggir dari circle
data, features = make_circles(n_samples=100, shuffle=True, noise=None, random_state=None, factor=0.8)
plt.scatter(data[:,0], data[:,1])
plt.show
features
```

```python
# ini inisialisasi awal dari pusat cluster
centers = [(-2, -2), (-2, 1.5), (1.5, -2), (2, 1.5)]
data2, features2 = make_blobs(n_samples=200, centers=centers, n_features=2, cluster_std=0.8, shuffle=False, random_state=42)
fig, ax= plt.subplots()
plt.scatter(data2[:,0], data2[:,1])
# # ax.scatter(np.asarray(centers).transpose()[0],
#           np.asarray(centers).transpose()[1], marker="o", s=100)
plt.plot()
plt.show
```
